### first simple model(only using sqft lot)


```python
#predictive model
y1= df['price']
X1=df[['sqft_lot']]

X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1,
                                                    test_size=0.2,
                                                    random_state=42)
lr1= LinearRegression()
lr1.fit(X_train1,y_train1)
print(lr1.score(X_train1, y_train1))
print(lr1.coef_)
```

```python
lr1.score(X_test1, y_test1)
```

```python
#inferential model
X1 = sm.add_constant(X1)
model = sm.OLS(y1,X1)
results = model.fit()
results.params
results.summary()
```

### multiple linear regression model (3 variables)

```python

y2= df['price']
X2=df[['sqft_lot', 'bedrooms', 'bathrooms']]

X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2,
                                                    test_size=0.2,
                                                    random_state=42)
lr2= LinearRegression()
lr2.fit(X_train2,y_train2)
print(lr2.score(X_train2, y_train2))
print(lr2.coef_)
```

```python
lr2.score(X_test2, y_test2)
```

```python
X2 = sm.add_constant(X_train2)
model2ols = sm.OLS(y_train2,X2)
results = model2ols.fit()
results.params
results.summary()
```

Based on the R^2 i can tell this is a better model than the first.
